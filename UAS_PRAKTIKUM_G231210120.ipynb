```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data from CSV

data = pd.read_csv('file stock.csv')
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
# a. Distribusi data pada setiap atribut dengan histogram
plt.figure(figsize=(12, 6))
fig = plt.gcf()
fig.set_facecolor('#94F008')
plt.subplot(2, 2, 1)
sns.histplot(data=df,color='red', x='Purchase Price', kde=True)
plt.title('Purchase Price Distribution')

plt.subplot(2, 2, 2)
sns.histplot(data=df,color='red', x='Current Price', kde=True)
plt.title('Current Price Distribution')

plt.subplot(2, 2, 3)
sns.histplot(data=df,color='red', x='Gain/Loss', kde=True)
plt.title('Gain/Loss Distribution')

plt.tight_layout()
plt.show()
```

```python
from tabulate import tabulate

top_dividend = df.nlargest(5, 'Dividend/share')[['Stock Name', 'Symbol', 'Dividend/share']]
table = tabulate(top_dividend, headers='keys', tablefmt='fancy_grid')
print("Top 5 Stocks with Highest Dividend/Share:")
print(table)
```

```python
from tabulate import tabulate

top_market_value = df.nlargest(5, 'Market Value')[['Stock Name', 'Symbol', 'Market Value']]
print("\nTop 5 Stocks with Highest Market Value:")
print(tabulate(top_market_value, headers='keys', tablefmt='psql'))
```

```python
from tabulate import tabulate

# d. Atribut Stock Name dan Symbol dengan Annual Yield tertinggi
top_annual_yield = df.nlargest(7, 'Annual Yield')[['Stock Name', 'Symbol', 'Annual Yield']]

# Membuat tabel dengan menggunakan tabulate
table = tabulate(top_annual_yield, headers='keys', tablefmt='fancy_grid')

# Menampilkan tabel
print("\nTop 7 Stocks with Highest Annual Yield:")
print(table)
```
